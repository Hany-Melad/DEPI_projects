# check data if number of left eye images  = right eye images 

```python
import scipy.io as sio
import glob
import os

def check_left_right_balance(root_dir):
    """
          left  right  .mat file.
    
    root_dir:   ( 'MPIIGaze/Data/Normalized')
    """
    mat_files = sorted(glob.glob(f'{root_dir}/**/day*.mat', recursive=True))
    print(f"  {len(mat_files)}  .mat...")
    
    all_balanced = True  # flag  
    
    for mat_path in mat_files:
        #   /   ( p00/day01)
        rel_path = os.path.relpath(mat_path, root_dir)
        print(f"\n : {rel_path}")
        
        #  
        mat = sio.loadmat(mat_path)
        data = mat['data'][0, 0]
        
        if 'left' not in data.dtype.names or 'right' not in data.dtype.names:
            print("     left  right!")
            all_balanced = False
            continue
        
        left_data = data['left'][0, 0]
        right_data = data['right'][0, 0]
        
        left_count = len(left_data['image'])
        right_count = len(right_data['image'])
        
        print(f"   - Left samples: {left_count}")
        print(f"   - Right samples: {right_count}")
        
        if left_count == right_count:
            print(" !")
        else:
            print(f"  ! (: {abs(left_count - right_count)})")
            all_balanced = False
    
    if all_balanced:
        print("\n    !")
    else:
        print("\n        .")

#  :
root_dir =  r'MPIIGaze/MPIIGaze/Data/Normalized'
check_left_right_balance(root_dir)
```

# check data if head pose of left and right eyes are equals to use average gaze and same head pose 

```python
import scipy.io as sio
import glob
import numpy as np
import random

def check_shared_pose(root_dir):
    """
       pose   right  left  .
     5 poses  .
    """
    #    .mat files   ( /)
    mat_files = sorted(glob.glob(f'{root_dir}/**/day*.mat', recursive=True))
    if not mat_files:
        raise ValueError("     .mat files  !")
    
    first_mat = mat_files[0]  #   ( p00/day01.mat)
    print(f"   : {first_mat}")
    
    #  
    mat = sio.loadmat(first_mat)
    data = mat['data'][0, 0]
    
    if 'right' not in data.dtype.names or 'left' not in data.dtype.names:
        raise ValueError("    'right'  'left'!")
    
    right_data = data['right'][0, 0]
    left_data = data['left'][0, 0]
    
    right_poses = right_data['pose']  # (N, 3)
    left_poses = left_data['pose']    # (N, 3)
    
    num_samples = len(right_poses)
    print(f"   : {num_samples}")
    
    if num_samples < 5:
        raise ValueError("    5    !")
    
    #  5 indices  (  )
    random.seed(42)  # 
    indices = random.sample(range(num_samples), 5)
    print(f" Indices : {indices}")
    
    all_match = True  # flag 
    
    for idx in indices:
        right_pose = right_poses[idx]
        left_pose = left_poses[idx]
        
        print(f"\n Index {idx}:")
        print("  Right Pose:", right_pose)
        print("  Left Pose: ", left_pose)
        
        #    (    floating point)
        if np.allclose(right_pose, left_pose, atol=1e-6):
            print("   !")
        else:
            print("    ! (  )")
            all_match = False
    
    if all_match:
        print("\n   Poses     Pose !")
    else:
        print("\n   Poses     .")

#  :
root_dir = r'MPIIGaze/MPIIGaze/Data/Normalized'  #    
check_shared_pose(root_dir)
```

# store data into touch file conatins images gaze pose for both right and left eyes

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import scipy.io as sio
import glob
import numpy as np
import os

# ============================================
# 1 DATASET  Pose (Independent Eyes + Pose Check)
# ============================================
class MPIIGazeDataset(Dataset):
    def __init__(self, root_dir, eyes=['right', 'left'], transform=None, check_poses=False):
        """
        root_dir: 'MPIIGaze/MPIIGaze/Data/Normalized'
        eyes: list of eye names to include, e.g., ['right', 'left']
        check_poses:  True    poses   left/right (  )
        """
        self.transform = transform
        self.images = []
        self.gazes = []
        self.poses = []
        self.eyes = eyes
        
        mat_files = sorted(glob.glob(f'{root_dir}/**/day*.mat', recursive=True))
        print(f"Loading {len(mat_files)} files...")
        
        if check_poses and len(eyes) == 2:
            self._check_poses(mat_files[0])  #     
        
        for f in mat_files:
            mat = sio.loadmat(f)
            data = mat['data'][0, 0]
            for eye in self.eyes:
                if eye not in data.dtype.names:
                    raise ValueError(f"Eye '{eye}' not found in {f}")
                else:
                    eye_data = data[eye][0, 0]
                    self.images.append(eye_data['image'])    # (N, 36, 60)
                    self.gazes.append(eye_data['gaze'])       # (N, 3)
                    self.poses.append(eye_data['pose'])       # (N, 3)
        
        self.images = np.concatenate(self.images, axis=0)  # (Total, 36, 60)
        self.gazes = np.concatenate(self.gazes, axis=0)     # (Total, 3)
        self.poses = np.concatenate(self.poses, axis=0)     # (Total, 3)
        
        print(f"Total samples: {len(self.images)}")
    
    def _check_poses(self, mat_path):
        """   poses   """
        print(f"  Poses : {mat_path}")
        mat = sio.loadmat(mat_path)
        data = mat['data'][0, 0]
        right_poses = data['right'][0, 0]['pose']
        left_poses = data['left'][0, 0]['pose']
        
        if len(right_poses) != len(left_poses):
            print("   Poses  !")
            return
        
        mismatches = 0
        for i in range(len(right_poses)):
            if not np.allclose(right_poses[i], left_poses[i], atol=1e-4):  #   
                mismatches += 1
        
        if mismatches > 0:
            print(f" {mismatches}  {len(right_poses)} Poses  ! (     )")
        else:
            print("   Poses !")
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        image = self.images[idx].astype(np.float32) / 255.0  # normalize to [0,1]
        image = torch.FloatTensor(image).unsqueeze(0)         # (1, 36, 60)
        gaze = torch.FloatTensor(self.gazes[idx])             # (3,)
        pose = torch.FloatTensor(self.poses[idx])             # (3,)
        
        if self.transform:
            image = self.transform(image)
        
        return image, gaze, pose


# ============================================
# 2   
# ============================================
def save_processed_dataset(root_dir, save_path, eyes=['right', 'left'], check_poses=False):
    print("="*60)
    print("    ...")
    print("="*60)
    
    dataset = MPIIGazeDataset(root_dir=root_dir, eyes=eyes, check_poses=check_poses)
    
    torch.save({
        'images': dataset.images,
        'gazes': dataset.gazes,
        'poses': dataset.poses,
        'metadata': {
            'total_samples': len(dataset),
            'image_shape': dataset.images.shape,
            'gaze_shape': dataset.gazes.shape,
            'pose_shape': dataset.poses.shape,
            'eyes': eyes
        }
    }, save_path)
    
    file_size = os.path.getsize(save_path) / (1024**2)
    
    print("\n" + "="*60)
    print("   !")
    print("="*60)
    print(f" : {save_path}")
    print(f" : {file_size:.2f} MB")
    print(f" : {len(dataset):,}")
    print(f"   : {dataset.images.shape}")
    print(f"   Gaze: {dataset.gazes.shape}")
    print(f"  Pose: {dataset.poses.shape}")
    print("="*60)


# ============================================
# 3  Train/Val 
# ============================================
def save_train_val_split(root_dir, train_path, val_path, eyes=['right', 'left'], split_ratio=0.8, check_poses=False):
    print("="*60)
    print("  ...")
    print("="*60)
    
    dataset = MPIIGazeDataset(root_dir=root_dir, eyes=eyes, check_poses=check_poses)
    
    train_size = int(split_ratio * len(dataset))
    indices = np.arange(len(dataset))
    np.random.seed(42)
    np.random.shuffle(indices)
    
    train_indices = indices[:train_size]
    val_indices = indices[train_size:]
    
    torch.save({
        'images': dataset.images[train_indices],
        'gazes': dataset.gazes[train_indices],
        'poses': dataset.poses[train_indices],
        'metadata': {
            'samples': len(train_indices),
            'split': 'train',
            'eyes': eyes
        }
    }, train_path)
    
    torch.save({
        'images': dataset.images[val_indices],
        'gazes': dataset.gazes[val_indices],
        'poses': dataset.poses[val_indices],
        'metadata': {
            'samples': len(val_indices),
            'split': 'validation',
            'eyes': eyes
        }
    }, val_path)
    
    train_size_mb = os.path.getsize(train_path) / (1024**2)
    val_size_mb = os.path.getsize(val_path) / (1024**2)
    
    print(f" Train: {len(train_indices):,} samples  {train_path} ({train_size_mb:.2f} MB)")
    print(f" Val: {len(val_indices):,} samples  {val_path} ({val_size_mb:.2f} MB)")


# ============================================
# 4  -  
# ============================================
if __name__ == "__main__":
    ROOT_DIR = r'MPIIGaze/MPIIGaze/Data/Normalized'
    
    #   1:       (  Poses)
    save_processed_dataset(
        root_dir=ROOT_DIR,
        save_path='mpii_gaze_full.pt',
        eyes=['right', 'left'],
        check_poses=True  #   
    )
    
    #   2:  Train/Val  ( )
    print("\n")
    save_train_val_split(
        root_dir=ROOT_DIR,
        train_path='mpii_gaze_train.pt',
        val_path='mpii_gaze_val.pt',
        eyes=['right', 'left'],
        split_ratio=0.8,
        check_poses=True
    )
```

# build random model to train data 
<p>
input:  (image+ pose vesctor )
<br>
output: (3d gaze vector )
</p>

```python
import cv2
import torch
import torch.nn as nn
import numpy as np



# ============================================
# 2 MODEL  ( Pose  Input + Output 3D Gaze Vector)
# ============================================
class GazeNetWithPose(nn.Module):
    def __init__(self):
        super(GazeNetWithPose, self).__init__()
        
        self.features = nn.Sequential(
            # Block 1: (1, 36, 60)  (32, 18, 30)
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.2),
            
            # Block 2: (32, 18, 30)  (64, 9, 15)
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.2),
            
            # Block 3: (64, 9, 15)  (128, 4, 7)
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(0.3),
        )
        
        #  regressor   3  pose
        flattened_size = 128 * 4 * 7  # 3584   features
        self.regressor = nn.Sequential(
            nn.Linear(flattened_size + 3, 256),  # +3  pose vector
            nn.ReLU(),
            nn.Dropout(0.4),
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 3),  # output: 3D gaze vector (x, y, z)
        )
    
    def forward(self, x, pose):
        #  features  
        x = self.features(x)
        x = torch.flatten(x, start_dim=1)  # (batch_size, 3584)
        
        #   pose   features
        x = torch.cat([x, pose], dim=1)    # (batch_size, 3584 + 3)
        
        #  regression  output
        x = self.regressor(x)
        return x  # (batch_size, 3) - 3D gaze vector
```

# Traning
<p> call reload class to read dataset from satored file  to use it into dataloader and split it into traning validation and testing </p>
<h4>tunable pramaters</h4>
<ul>
<li> epochs= 10
<li> leraning rate = lr=3e-3
<li> batch size = 512
<li> saved model = best_gaze_model.pth
</ul>

```python
import torch
from torch.utils.data import Dataset, DataLoader, random_split  #   random_split
import torch.nn as nn
import torch.optim as optim
import numpy as np


# ============================================
# 3 TRAINING SETUP
# ============================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using: {device}")

# Load data
#dataset = MPIIGazeDataset(r'MPIIGaze/MPIIGaze/Data/Normalized', eyes=['right', 'left'])

#load from saved file
dataset = PreloadedMPIIGaze(
        file_path='mpii_gaze_full.pt'  #   
    )
    



# Split: 80% train, 10% val, 10% test
total = len(dataset)
train_size = int(0.8 * total)
val_size = int(0.1 * total)
test_size = total - train_size - val_size

train_set, val_set, test_set = random_split(
    dataset, [train_size, val_size, test_size],
    generator=torch.Generator().manual_seed(42)
)

print(f"Train: {len(train_set)} | Val: {len(val_set)} | Test: {len(test_set)}")

train_loader = DataLoader(train_set, batch_size=512, shuffle=True, num_workers=0)
val_loader = DataLoader(val_set, batch_size=256, shuffle=False, num_workers=0)
test_loader = DataLoader(test_set, batch_size=256, shuffle=False, num_workers=0)

# Model, Loss, Optimizer
model = GazeNetWithPose().to(device)  #   
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=3e-3, weight_decay=1e-4) 
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)
# Count parameters
total_params = sum(p.numel() for p in model.parameters())
print(f"Model parameters: {total_params:,}")
print(model)

# ============================================
# 4 TRAINING LOOP
# ============================================
def angular_error(pred, target):
    """    3D vectors"""
    pred_np = pred.detach().cpu().numpy()
    target_np = target.detach().cpu().numpy()
    
    # Normalize vectors (unit vectors)
    pred_norm = np.linalg.norm(pred_np, axis=1, keepdims=True)
    target_norm = np.linalg.norm(target_np, axis=1, keepdims=True)
    pred_unit = pred_np / pred_norm
    target_unit = target_np / target_norm
    
    # Cosine similarity
    cos_sim = np.sum(pred_unit * target_unit, axis=1)
    cos_sim = np.clip(cos_sim, -1, 1)
    
    # Angular error in degrees
    angles = np.arccos(cos_sim) * 180 / np.pi
    return np.mean(angles)

EPOCHS = 10
best_val_error = float('inf')
train_losses = []
val_errors = []

for epoch in range(EPOCHS):
    # --- Train ---
    model.train()
    running_loss = 0
    for images, gazes, poses in train_loader:  #   poses
        images, gazes, poses = images.to(device), gazes.to(device), poses.to(device)
        
        optimizer.zero_grad()
        outputs = model(images, poses)  #  forward  poses
        loss = criterion(outputs, gazes)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * images.size(0)
    
    train_loss = running_loss / len(train_set)
    train_losses.append(train_loss)
    
    # --- Validate ---
    model.eval()
    all_preds, all_targets = [], []
    with torch.no_grad():
        for images, gazes, poses in val_loader:  #   poses
            images, gazes, poses = images.to(device), gazes.to(device), poses.to(device)
            outputs = model(images, poses)  #  forward  poses
            all_preds.append(outputs)
            all_targets.append(gazes)
    
    all_preds = torch.cat(all_preds)
    all_targets = torch.cat(all_targets)
    val_error = angular_error(all_preds, all_targets)
    val_errors.append(val_error)
    
    scheduler.step(val_error)
    
    # Save best
    if val_error < best_val_error:
        best_val_error = val_error
        torch.save(model.state_dict(), 'best_gaze_model.pth')
        marker = '  BEST'
    else:
        marker = ''
    
    print(f"Epoch [{epoch+1}/{EPOCHS}] Loss: {train_loss:.6f} | Val Error: {val_error:.2f}{marker}")
```

# <h4> using pose estimation from mediapipe and write pose in left uper corner </h4>

```python
import cv2
import torch
import torch.nn as nn
import numpy as np
import mediapipe  #    mediapipe  
from mediapipe.python.solutions import face_mesh as mp_face_mesh

# ============================================
# 1  (GazeNetWithPose)
# ============================================
class GazeNetWithPose(nn.Module):
    def __init__(self):
        super(GazeNetWithPose, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32), nn.ReLU(inplace=True),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32), nn.ReLU(inplace=True),
            nn.MaxPool2d(2), nn.Dropout2d(0.2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64), nn.ReLU(inplace=True),
            nn.MaxPool2d(2), nn.Dropout2d(0.2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128), nn.ReLU(inplace=True),
            nn.MaxPool2d(2), nn.Dropout2d(0.3),
        )
        flattened_size = 128 * 4 * 7
        self.regressor = nn.Sequential(
            nn.Linear(flattened_size + 3, 256),
            nn.ReLU(inplace=True), nn.Dropout(0.3),
            nn.Linear(256, 64),
            nn.ReLU(inplace=True), nn.Dropout(0.2),
            nn.Linear(64, 3),
        )

    def forward(self, x, pose):
        x = self.features(x)
        x = torch.flatten(x, start_dim=1)
        x = torch.cat([x, pose], dim=1)
        x = self.regressor(x)
        return x


# ============================================
# 2   (Draw Pose Axis + Label)
# ============================================
def gaze_to_label(pitch, yaw):
    P_TH, Y_TH = 0.10, 0.10
    vert = "Up" if pitch < -P_TH else ("Down" if pitch > P_TH else "Center")
    horiz = "Left" if yaw < -Y_TH else ("Right" if yaw > Y_TH else "Center")
    
    if vert == "Center" and horiz == "Center": return "Center"
    if vert == "Center": return horiz
    if horiz == "Center": return vert
    return f"{vert}-{horiz}"

def draw_pose_axis(img, rvec, tvec, cam_matrix, dist, axis_len=50):
    #      3D
    axis = np.float32([[axis_len,0,0], [0,axis_len,0], [0,0,axis_len]]).reshape(-1,3)
    
    #     2D
    imgpts, _ = cv2.projectPoints(axis, rvec, tvec, cam_matrix, dist)
    
    #   ( )
    nose_tip, _ = cv2.projectPoints(np.array([(0.0, 0.0, 0.0)]), rvec, tvec, cam_matrix, dist)
    nose_tip = tuple(nose_tip[0].ravel().astype(int))

    #  
    # X axis (Red)
    pt1 = tuple(imgpts[0].ravel().astype(int))
    cv2.line(img, nose_tip, pt1, (0,0,255), 2) 
    
    # Y axis (Green)
    pt2 = tuple(imgpts[1].ravel().astype(int))
    cv2.line(img, nose_tip, pt2, (0,255,0), 2)
    
    # Z axis (Blue) -   
    pt3 = tuple(imgpts[2].ravel().astype(int))
    cv2.line(img, nose_tip, pt3, (255,0,0), 2)

    return img


# ============================================
# 3  
# ============================================
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = GazeNetWithPose().to(device)
try:
    state_dict = torch.load('best_gaze_model.pth', map_location=device)
    model.load_state_dict(state_dict)
    print(" Model loaded successfully")
except:
    print(" Warning: 'best_gaze_model.pth' not found. Using random weights.")
model.eval()

eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

# 3D Model Points
model_points = np.array([
    [0.0, 0.0, 0.0],             # Nose tip
    [0.0, -330.0, -65.0],        # Chin
    [-225.0, 170.0, -135.0],     # Left eye left corner
    [225.0, 170.0, -135.0],      # Right eye right corner
    [-150.0, -150.0, -125.0],    # Left mouth corner
    [150.0, -150.0, -125.0]      # Right mouth corner
], dtype=np.float32)

cap = cv2.VideoCapture(0)
if not cap.isOpened(): raise SystemExit(" No camera")

# ============================================
# 4  
# ============================================
while True:
    ret, frame = cap.read()
    if not ret: break
    
    frame = cv2.flip(frame, 1)
    h, w, c = frame.shape
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Camera Matrix Update (Based on current frame size)
    focal_length = w
    center = (w/2, h/2)
    camera_matrix = np.array([[focal_length, 0, center[0]], [0, focal_length, center[1]], [0, 0, 1]], dtype=np.float32)
    dist_coeffs = np.zeros((4, 1))

    # 1.     Pose
    pose_vec = np.zeros(3)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb_frame)
    
    rvec, tvec = None, None

    if results.multi_face_landmarks:
        landmarks = results.multi_face_landmarks[0].landmark
        image_points = np.array([
            [landmarks[1].x * w, landmarks[1].y * h],    # Nose
            [landmarks[152].x * w, landmarks[152].y * h], # Chin
            [landmarks[226].x * w, landmarks[226].y * h], # Left Eye
            [landmarks[446].x * w, landmarks[446].y * h], # Right Eye
            [landmarks[57].x * w, landmarks[57].y * h],   # Left Mouth
            [landmarks[287].x * w, landmarks[287].y * h]  # Right Mouth
        ], dtype=np.float32)

        success, rvec, tvec = cv2.solvePnP(model_points, image_points, camera_matrix, dist_coeffs)
        
        if success:
            #     (Head Pose Visualization)
            draw_pose_axis(frame, rvec, tvec, camera_matrix, dist_coeffs)

            pose_vec = rvec.flatten()
            pose_norm = np.linalg.norm(pose_vec)
            pose_vec = pose_vec / pose_norm if pose_norm > 0 else pose_vec
            
            #    
            pose_text = f"Head: P={pose_vec[0]:.2f} Y={pose_vec[1]:.2f}"
            cv2.putText(frame, pose_text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)

    # 2.     Gaze
    eyes = eye_cascade.detectMultiScale(gray, 1.3, 5, minSize=(30, 30))
    
    if len(eyes) > 0:
        (x, y, ew, eh) = eyes[0]
        eye_img = cv2.resize(gray[y:y+eh, x:x+ew], (60, 36))
        
        #   Input
        eye_input = torch.from_numpy(eye_img.astype(np.float32)/255.0).unsqueeze(0).unsqueeze(0).to(device)
        pose_input = torch.from_numpy(pose_vec).unsqueeze(0).float().to(device)

        # 
        with torch.no_grad():
            gaze_3d = model(eye_input, pose_input)[0].cpu().numpy()

        #   Angles
        gaze_norm = np.linalg.norm(gaze_3d)
        if gaze_norm > 0: gaze_3d /= gaze_norm
        
        pitch = np.arcsin(-gaze_3d[1])
        yaw = np.arctan2(-gaze_3d[0], -gaze_3d[2])
        label = gaze_to_label(pitch, yaw)

        # 
        cv2.rectangle(frame, (x, y), (x+ew, y+eh), (0, 255, 0), 2)
        
        cx, cy = x + ew//2, y + eh//2
        length = 60
        dx = int(-length * np.sin(yaw))
        dy = int(-length * np.sin(pitch))
        
        #   
        cv2.arrowedLine(frame, (cx, cy), (cx+dx, cy+dy), (0, 0, 255), 2, tipLength=0.3)
        
        cv2.putText(frame, f"Gaze: {label}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    cv2.imshow('Gaze & Head Pose', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): break

cap.release()
cv2.destroyAllWindows()
face_mesh.close()
```
